```python
#Write a function cust_data() to ask user to enter their names and age to store data in customer.txt file.
def cust_data():
    name=input("Enter your name:")
    age=int(input("Enter your age:"))
    f=open("customer.txt","a+")
    f.write(name+"\n")
    f.write(str(age)+"\n")
cust_data()
```

```python
# Write a function count_lines() to count and display the total number of lines from the file. Consider the following lines for 
# the file – friends.txt.
# Friends are crazy, Friends are naughty !
# Friends are honest, Friends are  best !
# Friends are like keygen, friends are like license key !
# We are nothing without friends, Life is not possible without friends !
def count_lines():
    count=0
    f=open("friends.txt","a+")
    f.seek(0)
    for line in f:
        count+=1
    print("number of lines:",count)
    f.seek(0)
    print(f.read())
count_lines()
```

```python
# Write a function display_oddLines() to display odd number lines from the text file. Consider the following lines for the file – friends.txt.
# Friends are crazy, Friends are naughty !
# Friends are honest, Friends are  best !
# Friends are like keygen, friends are like license key !
# # We are nothing without friends, Life is not possible without friends 
def display_oddlines():
    f=open("friends.txt","r")
    line_no=1
    for line in f:
        if line_no%2!=0 :
            print(f.readline())
display_oddlines()
```

```python
# Write a Python program to reverse the content of a one file and store it in second file and also convert content of second 
# file into uppercase and store it in third file and also count number of Vowels in third file and also print only 2nd line from 
# the content of third file.
# Examples:
# If data file one contains the following data:
# Friends are crazy, Friends are naughty !
# Friends are honest, Friends are  best !
# Output 1:
# ! tseb  era sdneirF ,tsenoh era sdneirF
# ! ythguan era sdneirF ,yzarc era sdneirF
# Output 2:
# ! TSEB  ERA SDNEIRF ,TSENOH ERA SDNEIRF
# ! YTHGUAN ERA SDNEIRF ,YZARC ERA SDNEIRF
# Output 3:
# Vowels = 22
# Output 4:
# ! YTHGUAN ERA SDNEIRF ,YZARC ERA SDNEIRF
f=open("city.txt","r")
d=f.read()
rev=d[::-1]
f.close()
f1=open("a.txt","w")
f1.write(rev)
f1.close()

f2=open("a.txt","r")
d1=f2.read()
uprr=d1.upper()
f2.close()
f3=open("b.txt","w")
f3.write(uprr)

#count number of vowels
f4=open("b.txt","r")
lines=f4.readlines()

vcount=0
vowels='AEIOU'
for line in lines:
    for ch in line:
        if ch in vowels:
            vcount+=1
print(vcount)
if len(lines)>=2:
    print(f4.readline())
else:
    print("lines is less than 2 lines")
    f4.close()
```

```python
#Write a python program to extract a list of all four-letter words that start and end with the same letter from a given text 
#file 
f=open("b.txt","r")
d=f.read().split()
print(d)
count=0
for i in d:
    if len(i)==4 and i[0].lower()==i[-1].lower():
        count+=1
print(count)
```
